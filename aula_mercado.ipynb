```python
# o nosso desafio é mostrar os produtos de um mercado
# quero poder escolher os produtos, para ver infos
# quero poder adicionar novos produtos
# quero poder excluir um produto do mercado

#------------

# criar as variáveis globais para o software
lista_produtos= ['Detergente ypê, '
                'refrigerante Gengibirra, '
                'Arroz, '
                'Feijão, '
                'Agua mineral' ]

lista_preço =   [2.50 ,5,7.25,4,3.45]
```

```python
# para criarmos uma função
# para criarmos um apanhado de código, devemos
# armazenar eles dentro de um bloco e dar um nome

# função 1 - ver produtos
def ver_produtos():
  print(f'Os produtos que nós temos são:{lista_produtos} ')
  print(f'e seus preços são: {lista_preço}')

def adicionar_produtos():
  novo_produto = input('Informe o nome do produto')
  novo_preço = float(input ('Informe o preço: '))
  lista_produtos.append(novo_produto)
# /\ CHAMAMOS A LISTA - ATRIBUIMOS A FUNÇÃO APPEND (ADICIONA) -
  # COLOCAMOS O QUE VAI SER ADICIONADO

  lista_preço.append(novo_preço)
```

```python
print('-----MENU DO MERCADO -----')
print('Escolha a opcao:')
print('1- ver produtos')
print('2- cadastrar novo produto')
opcao = int(input('informe a opcao desejada'))



if opcao == 1:

  ver_produtos()

elif opcao == 2:
  adicionar_produtos()
  ver_produtos()
else:

  print('opcao não encontrada')
```
