```python
# import de bibliotecas
import pandas as pd
```

```python
# ler os três arquivos
jan = pd.read_csv('/content/vendas_janeiro.csv')
fev = pd.read_csv('/content/vendas_fevereiro.csv')
mar = pd.read_csv('/content/vendas_marco.csv')

print(jan)
```

```python
# transformar em dados
df_jan = pd.DataFrame(jan)
df_fev = pd.DataFrame(fev)
df_mar = pd.DataFrame(mar)

print(df_jan)
```

```python
# concatenar
df_trimestre_completo = pd.concat([df_jan,df_fev,df_mar], ignore_index=True)
```

```python
# criar o arquivo
df_trimestre_completo.to_csv('trimestre_completo.csv', index=False)
```

```python
# kpi
display(df_trimestre_completo)
```

```python
df_trimestre_completo['Faturamento Bruto por Item'] = df_trimestre_completo['Preço'] * df_trimestre_completo['Qtd']
faturamento_bruto_total = df_trimestre_completo['Faturamento Bruto por Item'].sum()

print(f"Faturamento Bruto Total do Trimestre: R$ {faturamento_bruto_total:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

faturamento_por_loja = df_trimestre_completo.groupby('Loja')['Faturamento Bruto por Item'].sum().reset_index()
faturamento_por_loja = faturamento_por_loja.sort_values(by='Faturamento Bruto por Item', ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x='Loja', y='Faturamento Bruto por Item', data=faturamento_por_loja, palette='viridis')
plt.title('Faturamento Bruto por Loja no Trimestre')
plt.xlabel('Loja')
plt.ylabel('Faturamento Bruto Total (R$)')
plt.ticklabel_format(style='plain', axis='y') # Remove scientific notation on y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
