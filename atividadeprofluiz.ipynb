```python
import requests
import pandas as pd

# Coordenadas (Sumaré - SP)
lat = -22.82
lon = -47.27

# Quantidade de dias desejada
quantos_dias = 7

# URL da API
url = 'https://api.open-meteo.com/v1/forecast'

# Parâmetros
parametros = {
    'latitude': lat,
    'longitude': lon,
    'daily': ['temperature_2m_max', 'precipitation_sum'],
    'timezone': 'auto',
    'past_days': quantos_dias
}

# Requisição
resposta = requests.get(url, params=parametros)
dados = resposta.json()

# Dados diários
dados_diarios = dados['daily']

# Criando DataFrame
df_dados_clima = pd.DataFrame(dados_diarios)

# Renomeando colunas
df_dados_clima.columns = ['data', 'temp_max', 'chuva_mm']

# Garantindo formato de data e pegando APENAS os últimos 7 dias
df_dados_clima['data'] = pd.to_datetime(df_dados_clima['data'])
df_dados_clima = df_dados_clima.sort_values('data').tail(7)

# Salvando CSV
df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)

# Exibindo resultado
print(df_dados_clima)
```
